 이미지 전처리 및 이진화, 침식, 팽창 연산을 수행하는 과정입니다


*   이미지 이진화: 밝고 어두운 영역을 분리합니다.
*   침식과 팽창 연산:
-침식(Erosion): 밝은 영역 축소, 노이즈 제거.
-팽창(Dilation): 밝은 영역 확장, 구멍 채우기.



*  소금-후추 노이즈 처리:
팽창-침식 연산을 결합하여 노이즈 제거 및 경계선 개선을 시도합니다.
*   다양한 이미지를 사용하여 모폴로지 연산의 효과를 시각적으로 확인합니다.





```python
img = cv2.imread('/content/gdrive/MyDrive/Colab Notebooks/디지털영상처리/milkdrop.bmp')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gimg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gimg.shape)
#plt.imshow(gimg1, cmap='gray')
cv2_imshow(gimg)
```

```python
gBin = np.zeros((gimg.shape[0], gimg.shape[1]), dtype = np.ubyte)

threshold = 150

for h in range(gimg.shape[0]):
  for w in range(gimg.shape[1]):
    if(gimg[h,w]> threshold):
      gBin[h,w] = 255

cv2_imshow(gBin)
```

```python
kernel = np.ones((3,3), np.uint8)
erodeimg = cv2.erode(gBin, kernel, iterations=5)
cv2_imshow(erodeimg)
```

```python
kernel = np.ones((3,3), np.uint8)
dilateimg = cv2.dilate(gBin, kernel, iterations=5)
cv2_imshow(dilateimg)

```

침식, 팽창, 이진화와 같은 기본적인 이미지 처리 기법을 사용하여 이미지를 처리하는 방법을 설명합니다.

```python
img = cv2.imread('/content/gdrive/MyDrive/Colab Notebooks/디지털영상처리/rice.bmp')
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

gimg = cv2.cvtColor(RGB_img, cv2.COLOR_RGB2GRAY)
print(gimg.shape)
#plt.imshow(gimg1, cmap='gray')
cv2_imshow(gimg)
```

```python
kernel = np.ones((3,3), np.uint8)
erodeimg = cv2.erode(gimg, kernel, iterations=1)
cv2_imshow(erodeimg)
```

```python
gBin = np.zeros((gimg.shape[0], gimg.shape[1]), dtype = np.ubyte)
```

```python
threshold = 140

for h in range(gimg.shape[0]):
  for w in range(gimg.shape[1]):
    if(gimg[h,w] > threshold):
      gBin[h,w] = 255

cv2_imshow(gBin)
```

```python

kernel = np.ones((3,3), np.uint8)
erodeimg = cv2.erode(gBin, kernel, iterations=1)
cv2_imshow(erodeimg)
```

```python
kernel = np.ones((3,1), np.uint8)
dilateimg = cv2.dilate(erodeimg, kernel, iterations=1)
cv2_imshow(dilateimg)
```

```python
kernel = np.ones((3,3), np.uint8)
openimg = cv2.morphologyEx(gBin, cv2.MORPH_OPEN, kernel)
cv2_imshow(openimg)
```
